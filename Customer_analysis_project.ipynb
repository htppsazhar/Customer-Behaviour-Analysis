### Treating Null Values

```python
df.isnull().sum()
```

```python
# Treating null values or replacing all values in the Electronics category 

df.loc[df['Category']=='Electronics','Size']='Not Applicable'
```

```python
df.groupby('Category')['Size'].unique()
```

### Size

```python
df['Size'].unique()
```

```python
# Treating null values or replacing all values in the clothing category 

df.loc[(df['Category']=='Clothing') & (df['Size'].isnull()), 'Size'] = df[df['Category']=='Clothing']['Size'].mode()[0]
```

```python
df.isnull().sum()
```

```python
df[df['Category']=='Clothing']['Size'].unique()
```

```python
df['Size'].unique()
```

```python
df[df['Category']=='Clothing']['Size'].value_counts(dropna=False)
```

```python
df[df['Size'].isnull()]['Category'].value_counts()
```

```python
df[df['Category']=='Accessories']['Size'].unique()
```

```python
df[df['Category']=='Footwear']['Size'].unique()
```

```python
mode_accessories = df.loc[df['Category']=='Accessories', 'Size'].mode()[0]

df.loc[(df['Category']=='Accessories') & (df['Size'].isnull()), 'Size'] = mode_accessories
```

```python
mode_footwear = df.loc[df['Category']=='Footwear', 'Size'].mode()[0]

df.loc[(df['Category']=='Footwear') & (df['Size'].isnull()), 'Size'] = mode_footwear
```

```python
# Treating null values present in the Review Rating 

df['Review Rating'].mean()
```

```python
df['Review Rating'].mode()
```

```python
df['Review Rating'].median()
```

```python
df.head(5)
```

```python
df.groupby('Item Purchased')['Review Rating'].mean()
```

```python
# Filling null values of Review Ratings with the mean of each Item purchased 

df['Review Rating']=df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform ('mean'))
```

```python
df['Review Rating'] = df['Review Rating'].round(2)
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.groupby('Item Purchased')['Review Rating'].mean()
```

### Previous Purchase

```python
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# filling null values of Purchase Amount with the mean of each Item Purchased

df.groupby('Item Purchased')['Purchase Amount (USD)'].mean()
```

```python
df['Purchase Amount (USD)']=df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform ('mean'))
```

```python
df.isnull().sum()
```

### Renaming the Headers

```python
df.columns = df.columns.str.replace("_", "")
```

```python
df.columns = df.columns.str.strip()  # remove leading/trailing spaces
df.columns = df.columns.str.replace(r"\s+", "_", regex=True)  # replace multiple spaces with one underscore
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = [
    'customer_id', 'age', 'gender', 'item_purchased', 'category',
    'purchase_amount_usd', 'location', 'size', 'color', 'season',
    'review_rating', 'subscription_status', 'shipping_type', 'discount_applied',
    'previous_purchases', 'payment_method', 'frequency_of_purchases'
]
```

```python
df.columns
```

```python
# replacing purchase_amount_(usd) to purchase_amount
df=df.rename(columns={'purchase_amount_usd':'purchase_amount'})
```
